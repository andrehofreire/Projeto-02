```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h1>Exploração e Preparação do Dataset</h1>

```python
#Realizando a leitura do arquivo csv.
df = pd.read_csv('athlete_events.csv')
```

```python
#Quantidade de linhas e colunas contidas no dataset.
df.shape
```

```python
#Utilizando a função head do pacote pandas para mostrar as primeiras 10 linhas do Dataframe.
df.head(5)
```

```python
#Renomeando colunas.
df.columns
```

```python
df.columns = ['ID', 'Nome', 'Genêro', 'Idade', 'Altura', 'Peso', 'País', 'Sigla', 'Jogos', 'Ano', 'Temporada', 'Cidade', 'Esporte', 'Evento', 'Medalha']
```

```python
#Tratando valores nulos
df.info()
```

```python
df.isnull().sum()
```

```python
df['Idade'] = df['Idade'].fillna(df['Idade'].mean())
df['Altura'] = df['Altura'].fillna(df['Altura'].mean())
df['Peso'] = df['Peso'].fillna(df['Peso'].mean())
```

```python
df.isnull().sum()
```

<h1><b>Visualizando os dados</b>

Quantidade de atletas ao longo do tempo

```python
verao = df[df["Temporada"] == "Summer"]
inverno = df[df["Temporada"] == "Winter"]
```

```python
verao = verao.groupby(["Ano"])["ID"].nunique().reset_index()
inverno = inverno.groupby(["Ano"])["ID"].nunique().reset_index()

fig = plt.figure(figsize=(15,12))
plt.subplot(212)
ax = sns.pointplot(x = verao["Ano"] , y = verao["ID"],color="red",markers="h")
ax.set_facecolor("White")
plt.grid(True,alpha=.2, color="Black")
plt.ylabel("Contagem")
plt.title("Olímpiadas de Verão - Numero de atletas ao longo dos anos",color="black")

fig = plt.figure(figsize=(15,12))
plt.subplot(212)
ax = sns.pointplot(x = inverno["Ano"] , y = inverno["ID"],color="blue",markers="h")
ax.set_facecolor("White")
plt.grid(True,alpha=.2, color="Black")
plt.ylabel("Contagem")
plt.title("Olímpiadas de Inverno - Numero de atletas ao longo dos anos",color="black")
```

Número de esportes ao longo do tempo

```python
espcont = df.groupby(['Ano','Temporada'])['Esporte'].nunique().reset_index()
plt.figure(figsize=(16,8))
sns.barplot("Ano", "Esporte", data=espcont, hue = 'Temporada')
```

Quantidade de Homens e Mulheres ao longo do tempo

```python
verao2 = df[df["Temporada"] == "Summer"]
inverno2 = df[df['Temporada'] == 'Winter']
```

```python
verao2 = verao2.groupby(['Ano', 'Genêro'])['ID'].nunique().reset_index()
inverno2 = inverno2.groupby(['Ano', 'Genêro'])['ID'].nunique().reset_index()

fig = plt.figure(figsize=(15,12))
plt.subplot(211)
ax = sns.pointplot(x = verao2["Ano"] , y = verao2["ID"],
                   markers="h" , hue = verao2["Genêro"],palette = ["blue","red"])
plt.grid(True)
plt.ylabel("Quantidade")
plt.title("Olímpiadas de Verão - Número de homens e mulheres ao longo dos anos",color="b")

fig = plt.figure(figsize=(15,12))
plt.subplot(211)
ax = sns.pointplot(x = inverno2["Ano"] , y = inverno2["ID"],
                   markers="h" , hue = inverno2["Genêro"], palette = ['red', 'blue'])
plt.grid(True)
plt.ylabel("Quantidade")
plt.title("Olímpiadas de Inverno - Número de homens e mulheres ao longo dos anos",color="b")
```

```python
def plot_corr(df, size=10):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)  
    plt.xticks(range(len(corr.columns)), corr.columns) 
    plt.yticks(range(len(corr.columns)), corr.columns)  
```

```python
plot_corr(df)
```
